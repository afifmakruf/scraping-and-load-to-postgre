```python
# Import library webdriver, Beautifulsoup, Pandas
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

# Extract

- Web Scraping

Tentukan data yang ingin diambil dan jalankan driver (bot) untuk membuka dan mengambil data dari web yang dituju, simpan di dalam soup

```python
driver = webdriver.Chrome() # Jalankan driver
# Membuat list kosong tiap kolom dan mendefinisikan nama tiap kolom
product        = []
price          = []
sold           = []
rating         = []
image          = []
link           = []       

for page in range(1,12):    # Menentukan range halaman yang ditentukan
    driver.get(f'https://ottencoffee.co.id/categories/coffee?page={page}') # Menjadikan page semula menjadi range page yang diinginkan
    driver.maximize_window() # Membuat jendela popup maximize
    scheight = .1  
    while scheight < 9.9:    # Membiarkan driver secara otomatis melakukan scroll agar data terbaca satu halaman full dengan while loop
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight/%s);" % scheight)
            scheight += .0015    

    html = driver.page_source   # Menentukan bahwa html adalah page_source dari driver
    soup = BeautifulSoup(html, 'html.parser')   # Menjadikan soup sebagai data yang berhasil diambil
    # Scraping data dari tokopedia seblak
    for data in soup.find_all('div', {'class':'sc-drKuOJ jueqIz'}): # Menentukan kotak per produk
        try:
            product.append(data.find('div', {'class':'sc-fnwBNb gqOztv title-product'}).get_text())
        except:
            product.append(None)
        try:
            price.append(data.find('div', {'class':'sc-dnqmqq iSQPAv'}).get_text())
        except:
            price.append(None)
        try:
            sold.append(data.find('span', {'class':'sc-bZQynM licOZQ eo-color-gray gsb-flex center-text'}).get_text())
        except:
            sold.append(None)
        try:
            rating.append(data.find('div', {'class':'gsb-flex'}).get_text())
        except:
            rating.append(None)
        try:
            image.append(data.find('img', {'class':'sc-cugefK bhRvD product-img'})['src'])
        except:
            image.append(None)
        try:
            link.append(data.find('a')['href'])
        except:
            link.append(None)
driver.close()  # Tutup driver
```

Berdasarkan data di atas, pada kolom sold, terdapat data yang tertulis masih mengandung (rb), sehingga untuk menganalisa data belum dapat dilakukan dengan baik, sehingga tulisan rb akan diganti dengan angka 000, dan juga 00

```python
# Fungsi untuk mengonversi format 'rb' atau angka biasa ke angka
def convert_to_number(text):
    if not text:  # Jika nilai None atau kosong, kembalikan None
        return None
    # Ambil angka sebelum "Terjual"
    text = text.split()[0]
    
    if "rb" in text:  # Jika mengandung 'rb'
        text = text.replace("rb", "").strip()  # Hapus 'rb'
        text = text.replace(",", ".")  # Ganti koma dengan titik
        if "." in text:
            return float(text) * 1000  # Jika desimal
        else:
            return int(text) * 1000  # Jika bilangan bulat
    else:  # Jika tidak mengandung 'rb', langsung konversi ke integer
        return int(text)

# Terapkan fungsi ke kolom 'Penjualan'
df["Sold"] = df["Sold"].apply(
    lambda x: convert_to_number(x) if isinstance(x, str) else None
)
```
